```python
def check_positions(squad):
    dct = {'Midfielder': 0,
           'Defender': 0,
           'Forward': 0,
           'GoalKeeper': 0}
    for player in squad:
        dct[player[1]] += 1
    if dct['Midfielder'] != 5:
        return False
    if dct['Defender'] != 5:
        return False
    if dct['Forward'] != 3:
        return False
    if dct['GoalKeeper'] != 2:
        return False
    return True
        
```

```python
def count_points(squad):
    total = 0
    points = [a[2] for a in squad]
    points = sorted(points, reverse=True)
    total = sum(points[:11])
    return total
```

```python
def calculate_cost(squad):
    player_costs = [p[3] for p in squad]
    return sum(player_costs)
```

```python
Defenders = [p for p in player_predictions if p[1] == 'Defender']
Midfielders = [p for p in player_predictions if p[1] == 'Midfielder']
Forwards = [p for p in player_predictions if p[1] == 'Forward']
GoalKeepers = [p for p in player_predictions if p[1] == 'GoalKeeper']
```

```python
Defender_Combos = itertools.combinations(Defenders, 5)
Midfielder_Combos = itertools.combinations(Midfielders, 5)
Forward_Combos = itertools.combinations(Forwards, 3)
GoalKeeper_Combos = itertools.combinations(GoalKeepers, 2)
```

```python
def count_points(squad):
    total = 0
    points = [a[2] for a in squad]
    return sum(points)
```

```python
from collections import defaultdict
```

```python
goalkeeper_dct = defaultdict(lambda: [0])
for players in GoalKeeper_Combos:
    total_cost = calculate_cost(players)
    total_points = count_points(players)
    if goalkeeper_dct[total_cost][0] < total_points:
        goalkeeper_dct[total_cost] = (total_points, players)
```

```python
forwards_dct = defaultdict(lambda: [0])
for players in Forward_Combos:
    total_cost = calculate_cost(players)
    total_points = count_points(players)
    if forwards_dct[total_cost][0] < total_points:
        forwards_dct[total_cost] = (total_points, players)
```

```python
defender_dct = defaultdict(lambda: [0])
for players in Defender_Combos:
    total_cost = calculate_cost(players)
    total_points = count_points(players)
    if defender_dct[total_cost][0] < total_points:
        defender_dct[total_cost] = (total_points, players)
```

```python
midfielder_dct = defaultdict(lambda: [0])
for players in Midfielder_Combos:
    total_cost = calculate_cost(players)
    total_points = count_points(players)
    if midfielder_dct[total_cost][0] < total_points:
        midfielder_dct[total_cost] = (total_points, players)
```

```python
goalkeepers_sorted_costs = sorted(goalkeeper_dct, key=lambda x: goalkeeper_dct[x][0], reverse=True)
defenders_sorted_costs = sorted(defender_dct, key=lambda x: defender_dct[x][0], reverse=True)
forwards_sorted_costs = sorted(forwards_dct, key=lambda x: forwards_dct[x][0], reverse=True)
midfielders_sorted_costs = sorted(midfielder_dct, key=lambda x: midfielder_dct[x][0], reverse=True)
```

```python
sum([goalkeepers_sorted_costs[1], defenders_sorted_costs[1], midfielders_sorted_costs[1], forwards_sorted_costs[1]])
```

```python
def count_points(squad):
    total = 0
    points = [a[2] for a in squad]
    points = sorted(points, reverse=True)
    total = sum(points[:11])
    return total
```

```python
%%time
best_points = 0
best_squad = []
bank = 960
for g in goalkeepers_sorted_costs:
    for d in defenders_sorted_costs:
        for f in forwards_sorted_costs:
            for m in midfielders_sorted_costs:
                total = g + d + m + f
                if total <= bank:
                    squad = goalkeeper_dct[g][1] + defender_dct[d][1] + midfielder_dct[m][1] + forwards_dct[f][1]
                    points = count_points(squad)
                    if points > best_points:
                        best_points = points
                        best_squad = squad
                        
            
```
