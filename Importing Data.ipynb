```python
def extract_player_links(results, clear=False):
    if not hasattr(extract_player_links, 'dct'):
        setattr(extract_player_links, 'dct', {})
    
    if clear:
        extract_player_links.dct.clear()
    
    for link in results:
        found = re.search(string=link, pattern=r'players/(\d+)/(\S+)/')
        if not found:
            continue
        player_id = found.group(1)
        player_name = parse_unicode(found.group(2))
        extract_player_links.dct[player_id] = player_name

```

```python
from urllib.parse import unquote

def parse_unicode(string):
    try:
        return unquote(string, errors='strict')
    except UnicodeDecodeError:
        return unquote(string, encoding='latin-1')
```

```python
def get_PL_stats(player_id, player_name):

    import requests
    import time
    
    time.sleep(1)  # Prevent accidently bombarding PL with requests

    stats_url    = "https://www.premierleague.com/players/{}/player/stats".format(player_id)
    overview_url = "https://www.premierleague.com/players/{}/player/overview".format(player_id)
    
    stats_page = requests.get(stats_url).text
    overview_page = requests.get(overview_url).text
    
    return {'stats': stats_page, 'overview': overview_page}
    
```

```python
pages = get_PL_stats(3512, 'Alex-Smith')
```

```python
def extract_PL_stats(stats, overview):
    
    from bs4 import BeautifulSoup
    
    soup_1 = BeautifulSoup(stats, 'html.parser')
    soup_2 = BeautifulSoup(overview, 'html.parser')
    
    
    return {'stats': soup_1, 'overview': soup_2}
    
```

```python
def extract_fields_from_stats(numeric_fields, soup):
    fields = {}
    
    page_text = soup.get_text()
    page_text = page_text.replace('\n', ' ')
    page_text = page_text.replace('\r', ' ')
    
    for field in numeric_fields:
        values = re.findall(pattern=r'{}\s+(\d+)'.format(field), flags=re.IGNORECASE, string=page_text)
        fields[field] = 0 if len(values) == 0 else int(values[0])
        
    fields['Team'] = re.findall(pattern=r'Club\s{4}(.+?)\s{2}Position', string=page_text)[0]

    return fields
        
```

```python
numeric_fields = ['Wins', 
                  'Goals Per Match', 
                  'Penalties Scored',
                  'Big Chances Missed', 
                  'Goals',
                  'Losses', 
                  'Appearances', 
                  'Clean Sheets', 
                  'Goals Conceded', 
                  'Own Goals', 
                  'Assists', 
                  'Big Chances Created', 
                  'Yellow Cards', 
                  'Red Cards']
```
